```python
#Load into Pandas
import pandas as pd
df = pd.read_csv("metadata.csv")
df
```

```python
#Examine the first few rows and structure
print(df.info())
```

```python
#Check DataFrame dimensions
print(df.shape)  # rows, columns
```

```python
#Identify data types
print(df.dtypes)
```

```python
#Check missing values
print(df.isnull().sum())
```

```python
#Generate summary statistics
print(df.describe())
```

```python
#Identify columns with many missing values
missing = df.isnull().mean().sort_values(ascending=False)
print(missing)
```

```python
#Drop unimportant columns with too many missing values.
df_clean = df.dropna(subset=["title", "publish_time"])
```

```python
#Count papers by year
papers_per_year = df_clean["year"].value_counts().sort_index()
```

```python
from collections import Counter
import re

words = " ".join(df_clean["title"].dropna()).lower()
words = re.findall(r'\b\w+\b', words)
common_words = Counter(words).most_common(20)
print(common_words)
```
